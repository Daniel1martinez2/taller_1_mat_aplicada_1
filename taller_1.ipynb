```python
import numpy.random as random

def f(x):
	if (x < -3):
		return x+2
	elif (-3<=x<0):
		return -x
	elif (x>=0):
		return 2

x_values = random.randint(-1000, 1000, 100)

results_func = [f(x) for x in x_values]
```

```python
import matplotlib.pyplot as plt

plt.plot(x_values, results_func, '.', color='red', markersize=5)
plt.show()
```

### 2.2
- _a_ y _b_ deben poder ser elegidos por el usuario usando la función Input



```python
a = int(input("Ingrese el valor de a: "))
b = int(input("Ingrese el valor de b: "))

def f(x):
	if (x < 0):
		return a-(x**2)
	elif (0<=x<=1):
		return 2
	elif (x>1):
		return (x**2) - 2*x + b

x_values = random.randint(-1000, 1000, 100)

results_func = [f(x) for x in x_values]

plt.plot(x_values, results_func, '.', color='blue', markersize=5)
plt.title(f"a={a}, b={b}")
plt.show()
```

```python
def operaciones (numeros):
	cuadrados = [num**2 for num in numeros]
	cubos = [num**3 for num in numeros]
	raices = [num**0.5 for num in numeros]

	return cuadrados, cubos, raices


mi_lista = [2, 4, 6, 8, 10]
resultado = operaciones (mi_lista)

print("Cuadrados:", resultado[0])
print("Cubos:", resultado[1])
print("Raíces:", resultado[2])
```
