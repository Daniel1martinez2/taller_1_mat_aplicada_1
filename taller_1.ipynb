```python
import numpy.random as random

def f(x):
	if (x < -3):
		return x+2
	elif (-3<=x<0):
		return -x
	elif (x>=0):
		return 2

x_values = random.randint(-1000, 1000, 100)

results_func = [f(x) for x in x_values]


```

```python
import matplotlib.pyplot as plt

plt.plot(x_values, results_func, '.', color='red', markersize=5)
plt.show()
```
